```python
import pandas as pd
from shutil import copyfile
```

```python
df = pd.read_csv('data/train_no_unknown_df.csv')
df.head()
```

```python
for i in range (len(df)):
    # Se obtiene el nombre de la foto
    filename = df.iloc[i, 0] + '.jpg'
    
    # Se crean las rutas de src y dest
    src = 'data/jpeg/train/' + filename
    dst = 'data/jpeg_no_unknown/train_no_unknown/' + filename
    
    # Se copia
    copyfile(src, dst)
    
    print("Copiada foto %i de %i" % (i+1, len(df)))
```
